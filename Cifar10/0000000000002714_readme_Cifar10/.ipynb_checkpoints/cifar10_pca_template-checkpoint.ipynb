```python
import cifar10
import matplotlib.pyplot as plt
```

```python
cifar10.data_path = "data/CIFAR-10/"
```

```python
cifar10.maybe_download_and_extract()
```

```python
class_names = cifar10.load_class_names()
class_names
```

```python
images_train, cls_train, labels_train = cifar10.load_training_data()
images_test, cls_test, labels_test = cifar10.load_test_data()
```

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
images_train.shape, images_test.shape
```

```python
cls_train, cls_test
```

```python
labels_train, labels_test
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1)
    ax.imshow(images_train[i], cmap=plt.cm.bone)
plt.show()
```

```python
# PCA without reduction to get best possible n_components

# 2d images
images_no, images_x, images_y, images_z = images_train.shape
images_2d = images_train.reshape(images_no, images_x*images_y*images_z)
images_2d.shape
```

```python
pca = PCA()
pca.fit(images_2d)
```

```python
# Getting best value for n_components
k = 0
total = 0
while total < 0.99:
    total += pca.explained_variance_ratio_[k]
    k += 1
k
```

```python
pca_cifar = PCA(n_components=k)
images_2d_transformed = pca_cifar.fit_transform(images_2d)
images_2d_transformed.shape
```

```python
# Regeneration of approximate images
images_2d_approx = pca_cifar.inverse_transform(images_2d_transformed)
images_2d_approx.shape
```

```python
images_approx = images_2d_approx.reshape(-1, 32, 32, 3)
images_approx.shape
```

```python
# Plotting approx images
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1)
    ax.imshow(images_approx[i], cmap=plt.cm.bone)
plt.show()
```

```python
# Reshaping test data
images_no, images_x, images_y, images_z = images_test.shape
images_2d_test = images_test.reshape(images_no, images_x*images_y*images_z)
images_2d_test.shape
```

```python
# Transforming test data
images_2d_test = pca_cifar.transform(images_2d_test)
images_2d_test.shape
```

# Using Random Forest

```python
clf1 = RandomForestClassifier()
clf1
```

```python
clf1.fit(images_2d_transformed, cls_train)
y_pred1 = clf1.predict(images_2d_test)
print(classification_report(cls_test, y_pred1))
print(confusion_matrix(cls_test, y_pred1))
print('Score: ', clf1.score(images_2d_test, cls_test))
```

# Using Logistic Regression

```python
clf2 = LogisticRegression()
```

```python
clf2
```

```python
clf2.fit(images_2d_transformed, cls_train)
y_pred2 = clf2.predict(images_2d_test)
print(classification_report(cls_test, y_pred2))
print(confusion_matrix(cls_test, y_pred2))
print('Score: ', clf2.score(images_2d_test, cls_test))
```

## Using Support Vector Machine

```python
clf3 = SVC(kernel='rbf', C=10, gamma=0.01)
clf3
```

```python
clf3.fit(images_2d_transformed, cls_train)
```

```python
y_pred3 = clf3.predict(images_2d_test)
print(classification_report(cls_test, y_pred3))
print(confusion_matrix(cls_test, y_pred3))
print('Score: ', clf3.score(images_2d_test, cls_test))
```
